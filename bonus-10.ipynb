```python
"""
Vrať se k práci se souborem temperature.csv, který obsahuje informace o průměrné teplotě v různých městech v listopadu 2017.

Vytvoř tabulku, která bude obsahovat údaje o teplotě za města Helsinki, Miami Beach a Tokyo.
Vytvoř krabicový graf a porovnej rozsah teplot v těchto městech.

"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
temperatures = "temperature.csv"
temperatures = pd.read_csv(temperatures)
temperatures.head()
```

```python
Helsinki = temperatures[
    (temperatures["City"] == "Helsinki")
    ]

Helsinki = pd.Series(Helsinki["AvgTemperature"])
Helsinki
```

```python
MiamiBeach = temperatures[
    (temperatures["City"] == "Miami Beach")
    ]

MiamiBeach = pd.Series(MiamiBeach["AvgTemperature"])
MiamiBeach
```

```python
Tokyo = temperatures[
    (temperatures["City"] == "Tokyo")
    ]

Tokyo = pd.Series(Tokyo["AvgTemperature"])
Tokyo
```

```python
TempTable = pd.DataFrame(columns= ["Helsinki", "Miami Beach", "Tokyo"])
TempTable["Helsinki"] = Helsinki
TempTable["Miami Beach"] = MiamiBeach
TempTable["Tokyo"] = Tokyo

TempTable

# Nevím, čím to je, že se mi nenačtou hodnoty do sloupců Miami Beach a Tokyo, napadá mě, jestli to není kvůli indexování?
```

```python
TempTable.plot(
    kind="box",
    title="Average Temperature Comparison",
    xlabel="City",
    ylabel="Average Temperature °F"
)
    
plt.show()
```
