```python
import pandas as pd
```

```python
platy = "platy_2021_02.csv"
platy = pd.read_csv(platy)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = [15,5]
```

```python
platy_Series = pd.Series(platy["plat"])
platy_Series
```

```python
platy_Series.hist(color="purple")
plt.title("Histogram výše mezd v Kč")
plt.xlabel("Mzdy v Kč")
plt.show()

# stackoverflow:    import numpy as np
#                   plt.hist(platy_Series, bins=np.arange(min(platy_Series), max(platy_Series), + 1, 1)) → nevím proč, ale nevygeneruje sloupce
```
